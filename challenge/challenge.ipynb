## import library

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
```

## ดึงข้อมูล Iris

```python
from sklearn.datasets import load_iris
iris = load_iris()
iris_data = iris.data
iris_target = iris.target
```

## แสดงข้อมูล 5 บรรทัดเเรก

```python
iris_df = pd.DataFrame(data=iris_data, columns=iris.feature_names)
iris_df['target'] = iris_target
iris_df.head()
```

```python
sns.scatterplot(data=iris_df, x='sepal length (cm)', y='sepal width (cm)', hue='target')
plt.show()
```

```python
iris_df.to_csv('iris_dataset.csv', index=False)
```

## Iris EDA

```python
# ดูข้อมูลเบื้องต้น
iris_df.info()
```

```python
# ดูสถิติพื้นฐาน
iris_df.describe()
```

```python
# ตรวจสอบค่าที่ขาดหายไป
iris_df.isnull().sum()
```

```python
# แสดง Histogram ของข้อมูล
sns.histplot(iris_df['sepal length (cm)'], kde=True)
plt.title('Distribution of Sepal Length')
plt.show()

# แสดง Pair Plot เพื่อดูความสัมพันธ์ระหว่างตัวแปร
sns.pairplot(iris_df, hue='target')
plt.show()
```

```python
# ตรวจสอบ Outliers ด้วย Boxplot
sns.boxplot(x=iris_df['sepal length (cm)'])
plt.title('Boxplot of Sepal Length')
plt.show()
```

```python
# คำนวณ correlation matrix
corr_matrix = iris_df.corr()

# แสดง correlation matrix
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Iris Dataset')
plt.show()
```

```python
# การแสดง Line Plot ของราคาหุ้น
nvidia['Close'].plot(title='NVIDIA Stock Price Over Time', figsize=(10, 6))
plt.xlabel('Date')
plt.ylabel('Close Price (USD)')
plt.show()

```

```python

# การแสดง Histogram ของราคาหุ้น
sns.histplot(nvidia['Close'], kde=True)
plt.title("Distribution of NVIDIA Stock Prices")
plt.show()
```
