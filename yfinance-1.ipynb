```python
!pip install yfinance
import yfinance as yf
```

```python
nvidia = yf.Ticker('NVDA')
nvidia_data = nvidia.history(period='1mo')  # ดึงข้อมูลหุ้นในช่วง 1 เดือนที่ผ่านมา
```

```python
nvidia_data.head()
```

```python
import matplotlib.pyplot as plt

nvidia_data['Close'].plot(title="NVIDIA Stock Price")
plt.show()
```

```python
nvidia_data.to_csv('nvidia_stock_data.csv')
```

```python
import yfinance as yf

# ดึงข้อมูลหุ้น NVIDIA ย้อนหลัง 6 เดือน
nvidia = yf.Ticker('NVDA')
nvidia_data = nvidia.history(period='6mo')  # period='6mo' หมายถึง 6 เดือนย้อนหลัง
```

```python
nvidia_data.to_csv('nvidia_stock_data_6months.csv')
```

```python
import matplotlib.pyplot as plt

nvidia_data['Close'].plot(title="NVIDIA Stock Price")
plt.show()
```
