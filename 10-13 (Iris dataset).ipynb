```python
from sklearn.datasets import load_iris
import pandas as pd

# โหลดข้อมูล
iris = load_iris()

# แปลงเป็น DataFrame
df_iris = pd.DataFrame(iris.data, columns=iris.feature_names)

# เพิ่มคอลัมน์ชื่อสายพันธุ์
df_iris["species"] = [iris.target_names[i] for i in iris.target]

# แสดงตัวอย่างข้อมูล
df_iris.head()
```

```python
# ขนาดของ dataset
print(f"Dataset shape: {df_iris.shape}")

# ดูข้อมูลตัวอย่าง
print(df_iris.head())

# ดูประเภทของข้อมูลแต่ละคอลัมน์
print(df_iris.info())

# ตรวจสอบค่าที่ซ้ำหรือค่าหายไป
print(df_iris.isnull().sum())
```

```python
# ดูค่าสถิติเบื้องต้น
df_iris.describe()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# ตั้งค่าขนาดกราฟ
plt.figure(figsize=(10, 6))

# วาด histogram
df_iris.hist(bins=20, figsize=(12, 8))
plt.suptitle("Histogram of Iris Dataset Features")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df_iris.iloc[:, :-1])
plt.title("Boxplot of Iris Dataset Features")
plt.show()
```

```python
import os

# ตรวจสอบว่ามีโฟลเดอร์ "data" หรือไม่ ถ้าไม่มีให้สร้างขึ้นมา
if not os.path.exists("data"):
    os.makedirs("data")

# บันทึก DataFrame เป็น CSV
df_iris.to_csv("Zbook Firefly 14 G8/iris_data.csv", index=False)

print("Exported iris_data.csv successfully!")
```
