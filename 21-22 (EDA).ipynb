```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

```python
# โหลดชุดข้อมูล
iris = load_iris()

# แปลงข้อมูลเป็น DataFrame
iris_data = pd.DataFrame(data=iris.data, columns=iris.feature_names)

# เพิ่มคอลัมน์ "species" เพื่อระบุชื่อของดอกไม้แต่ละชนิด
iris_data['species'] = iris.target
iris_data['species'] = iris_data['species'].replace({0: 'setosa', 1: 'versicolor', 2: 'virginica'})

# แสดงข้อมูล 5 แถวแรก
iris_data.head()
```

```python
print(iris_data.info())
```

```python
print(iris_data.describe())
```

```python
sns.boxplot(data=iris_data.drop(columns=['species']), orient="h")
plt.title("Boxplot ของคุณลักษณะในชุดข้อมูล Iris")
plt.show()
```

```python
sns.pairplot(iris_data, hue='species')
plt.suptitle("Pairplot ของคุณลักษณะในชุดข้อมูล Iris", y=1.02, fontsize=14)
plt.show()
```

```python
corr = iris_data.drop(columns=['species']).corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap ของชุดข้อมูล Iris")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import yfinance as yf
```

```python
# ดึงข้อมูลหุ้นของ NVIDIA ตั้งแต่ก่อตั้ง
nvidia = yf.Ticker('NVDA')
nvidia_data = nvidia.history(period='max')

# แสดงข้อมูล 5 แถวแรก
nvidia_data.head()
```

```python
print(nvidia_data.info())
```

```python
print(nvidia_data.describe())
```

```python
nvidia_data['Close'].plot(kind='hist', bins=50, figsize=(10, 6), alpha=0.7)
plt.title("Histogram ของราคาปิดหุ้น NVIDIA")
plt.xlabel("Close Price")
plt.ylabel("ความถี่")
plt.grid(True)
plt.show()
```

```python
nvidia_data['Close'].plot(figsize=(12, 6), title="แนวโน้มราคาปิดของ NVIDIA")
plt.xlabel("วันที่")
plt.ylabel("ราคาปิด")
plt.grid(True)
plt.show()
```

```python
corr = nvidia_data.corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Heatmap ของหุ้น NVIDIA")
plt.show()
```

```python
plt.scatter(nvidia_data['Volume'], nvidia_data['Close'], alpha=0.5)
plt.title("Volume vs Close Price ของ NVIDIA")
plt.xlabel("Volume")
plt.ylabel("Close Price")
plt.grid(True)
plt.show()
```

```python
max_close = nvidia_data['Close'].max()
min_close = nvidia_data['Close'].min()
avg_close = nvidia_data['Close'].mean()

print(f"ราคาปิดสูงสุด: {max_close}")
print(f"ราคาปิดต่ำสุด: {min_close}")
print(f"ราคาเฉลี่ย: {avg_close}")
```
